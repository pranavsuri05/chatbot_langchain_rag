```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    text = ""
    for page_num in range(pdf_document.page_count):
        page = pdf_document.load_page(page_num)
        page_text = page.get_text()
        print(f"Page {page_num + 1}:")
        print(page_text[:500])  # Print the first 500 characters of each page's text
        text += page_text
    return text

# Extract text from PDF and verify each page
file_path = "tsla-20231231-gen.pdf"
pdf_text = extract_text_from_pdf(file_path)
```

```python
from langchain.schema import Document
from langchain.text_splitter import CharacterTextSplitter

# Create Document objects from extracted PDF text
documents = [Document(page_content=pdf_text)]

# Split the text into chunks
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=100)
texts = text_splitter.split_documents(documents)

# Verify the chunked texts
print(texts[:3])  # Print the first 3 chunks to verify
```

```python
from langchain.text_splitter import CharacterTextSplitter

# Create Document objects from extracted PDF text
if pdf_text:
    documents = [Document(page_content=pdf_text)]
    text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=100)

    # Split the text into chunks
    texts = text_splitter.split_documents(documents)

    # Verify the chunked texts
    print(texts[:3])  # Print the first 3 chunks to verify
else:
    print("The PDF text is empty, extraction might have failed.")
```

```python
import os
from langchain.embeddings import OpenAIEmbeddings

# Set your OpenAI API key
os.environ["OPENAI_API_KEY"] = "apikey"

from langchain_openai import OpenAI
# Define the language model (OpenAI LLM)
llm = OpenAI(openai_api_key="apikey")   # Add your API key if it's not set globally

from langchain.vectorstores import FAISS  # Import FAISS
from langchain.chains import RetrievalQA

# Now initialize OpenAIEmbeddings
embeddings = OpenAIEmbeddings()
vector_store = FAISS.from_documents(texts, embeddings)
retriever = vector_store.as_retriever()
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever)
query = "Your test question here"
answer = qa_chain.invoke(query)
print(f"Answer: {answer}")
```

```python
# You can wrap this in a loop for multiple queries
print("Chatbot: Ask me anything (Type 'exit' to quit)")
while True:
    query = input("You: ")
    if query.lower() == 'exit':
        print("Chatbot: Goodbye!")
        break
    answer = qa_chain.invoke(query)
    print(f"Chatbot: {answer}")
```
